# Q2: Make dash that takes name input and outpt "Welcome to Data Science dashboard {name}"

```python
Q2 = JupyterDash()
name = ' '
inp1 = dcc.Input(id='my-input', value=name, type='text')
out1 = html.H2(id='my-H2', style={'color':'grey'})
Q2.layout = html.Div([inp1, out1])
@Q2.callback(
    Output('my-H2', 'children'),
    Input('my-input', 'value')
)
def update(name):
    return f'Welcome to Data Science dashboard {name}'
Q2.run_server(port = '2225')
```

# Q3: Make Plotly-Dash that takes two inputs (numbers) and output 3 numbers x, y , z= x+y
- hint: use html.Div, dcc.Input(type='number')

```python
Q3 = JupyterDash()
num1, num2 = 0, 0
inp1 = dcc.Input(id='my-input1', type='number')
inp2 = dcc.Input(id='my-input2', type='number')
out1 = html.H2(id='my-H2', style={'color':'grey'})
Q3.layout = html.Div([inp1,inp2, out1])
@Q3.callback(
    Output('my-H2', 'children'),
    Input('my-input1', 'value'),
    Input('my-input2', 'value')
)
def update(num1, num2):
    return f'{num1} + {num2} = {num1+num2}'
Q3.run_server(port = '2226')
```

# Q6: Make Plotly chart, then add it to a plotly_dash

```python
import plotly.express as px
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('titanic passenger list.csv')
df.head()
```

```python
pie = px.pie(df, values='pclass', names='pclass', title='Passenger Class Distribution')
pie.show()
```

```python
Q6 = JupyterDash()
Q6.layout = html.Div([
    dcc.Graph(figure=pie)
])
Q6.run_server(port='2230')
```

# Q7: From our data make plotly-dash with suitble chart, and Dropdown.

- Dropdown should contain at least 3 options, each one represents different column.
- Add label to dropdown.
- The chart data should change depending on the selected dropdown option.

```python
df.info()
```

```python
df.head()
```

```python
bar = px.histogram(df, x='sex',y='survived', color='pclass', barmode='group')
bar.show()
```

```python
Q7 = JupyterDash()
Q7.layout = html.Div([
    dcc.Dropdown(id='dropdown', 
                options= [{'label':i , 'value':i} for i in df['pclass'].value_counts().index.sort_values()]),
    html.Br(),
    dcc.Graph(id='graph')
])
@Q7.callback(
    Output('graph','figure'),
    Input('dropdown','value')
)
def update_figure(pclass):
    filtered_df = df[df['pclass'] == pclass]
    hist = px.histogram(filtered_df, x='sex',y='survived', color='pclass',
    barmode='group', title = f'survived people with class {pclass}')
    return hist
Q7.run_server(port='2231')
```

# Q8: add 2 more charts to the prevoius Ploty-Dash

```python
sct = px.scatter(df, x="pclass", y="age", color='survived', title="Scatter Plot")
sct.show()
```

```python
Q8 = JupyterDash()
Q8.layout = html.Div([
    dcc.Dropdown(id='dropdown', 
                options= [{'label':i , 'value':i} for i in df['pclass'].value_counts().index.sort_values()]),
    html.Br(),
    dcc.Graph(id='graph'),
    dcc.Graph(figure=pie),
    dcc.Graph(figure=sct)
])
@Q8.callback(
    Output('graph','figure'),
    Input('dropdown','value')
)
def update_figure(pclass):
    filtered_df = df[df['pclass'] == pclass]
    hist = px.histogram(filtered_df, x='sex',y='survived', color='pclass',
    barmode='group', title = f'survived people with class {pclass}')
    return hist
Q8.run_server(port='2231')
```
